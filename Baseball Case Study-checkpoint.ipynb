Removing Outliers

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z.shape
```

```python
threshold=3
print(np.where(z>3))
```

```python
df_new=df[(z<3).all(axis=1)]
print(df.shape)
print(df_new.shape)
```

```python
loss_percent=(30-29)/30*100
print(loss_percent)
```

```python
x=df_new.drop("W",axis=1)
y=df_new['W']
```

Linear Regresssion

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    print(f"At random state{i},the training accuracy is:-{r2_score(y_train,pred_train)}")
    print(f"At random state{i},the testing accuracy is:-{r2_score(y_test,pred_test)}")
    print("\n")
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=30)
```

```python
y_pred=lr.predict(x_train)
print("Training Accuracy:",lr.score(x_train,y_train)*100)
```

```python
print("Testing Accuracy:",lr.score(x_test,y_test)*100)
```

```python
print("Model Accuracy:",r2_score(y,lr.predict(x))*100)
```

Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_train)
```

```python
print("Training Accuracy:",reg.score(x_train,y_train)*100)
```

```python
print("Testing Accuracy:",reg.score(x_test,y_test)*100)
```

```python
print("Model Accuracy:",r2_score(y,reg.predict(x))*100)
```
